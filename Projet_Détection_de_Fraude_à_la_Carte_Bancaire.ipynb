```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='Class', data=df)
plt.title('Distribution of Fraud vs Legitimate Transactions')
plt.xticks([0, 1], ['Legitimate', 'Fraud'])
plt.show()
```

The dataset exhibits extreme class imbalance, with 284,315 legitimate transactions and only 492 fraudulent ones (approximately 578:1). This imbalance makes raw count visualizations potentially misleading, as the minority class is barely visible without proper scaling or annotation. Consequently, accuracy is not a reliable evaluation metric, and model performance should instead be assessed using metrics such as precision, recall, and F1-score to better capture the detection of rare fraud cases.

```python
plt.figure(figsize=(8,6))
sns.histplot(data=df, x='Amount', hue='Class', bins=50, kde=True)
plt.title('Distribution of Transaction Amount by Class')
plt.show()
```

The histplot with hue='Class' reveals that the vast majority of both legitimate and fraudulent transactions cluster at very low amounts (near $0–$100). However, legitimate transactions have a long heavy tail extending to thousands of dollars, while fraud transactions are concentrated in a much narrower, lower range. The KDE curves are nearly unreadable due to the scale difference caused by the class imbalance.

```python
sns.boxplot(x='Class', y='Amount', data=df)
plt.title('Transaction Amount by Class')
plt.ylim(0, 500)
plt.xticks([0, 1], ['Legitimate', 'Fraud'])
plt.show()
```

Fraudulent transactions tend to have a lower median amount and a more concentrated distribution compared to legitimate transactions. In contrast, legitimate transactions display a wider spread with a pronounced right skew due to high-value outliers. Despite these differences, there is substantial overlap between the two classes, indicating that transaction amount alone is not sufficient to reliably distinguish fraud. This feature may still contribute useful information when combined with other variables in a predictive model.

```python
import plotly.express as px

df['Hour'] = (df['Time'] // 3600).astype(int)

hourly_data = df.groupby('Hour')['Class'].value_counts().unstack(fill_value=0)
hourly_data.columns = ['Legitimate', 'Fraud']
hourly_data = hourly_data.reset_index()


fig = px.area(
    hourly_data,
    x='Hour',
    y=['Legitimate', 'Fraud'],
    title="When Do Fraudsters Attack? (Transaction Volume by Hour)",
    labels={'value': 'Number of Transactions', 'variable': 'Transaction Type'},
    color_discrete_map={'Legitimate': 'blue', 'Fraud': 'red'}
)


fig.update_yaxes(type="log")
fig.update_layout(template="plotly_white", height=500)
fig.show()
```

Fraud Hides in Low-Volume Shadows: The log-scale visualization reveals a critical behavioral divergence: legitimate transactions peak during business hours, whereas fraud maintains a low but persistent and stealthy footprint during the early morning hours (0-5 AM) . This temporal mismatch is the fraudster's tactical advantage—they operate when the system's signal-to-noise ratio is highest, betting on drowsy monitoring. The insight is clear: Rule-based scoring models must be time-aware, dynamically tightening thresholds during these nocturnal windows to catch the "hidden" attacks that volume-based alerts often miss.

```python


hourly_stats = df.groupby('Hour')['Class'].agg(['count', 'sum']).reset_index()
hourly_stats.columns = ['Hour', 'Total_Transactions', 'Fraud_Count']
hourly_stats['Fraud_Rate_Percent'] = (hourly_stats['Fraud_Count'] / hourly_stats['Total_Transactions']) * 100


fig = px.bar(
    hourly_stats,
    x='Hour',
    y='Fraud_Rate_Percent',
    title="The Danger Hours (Fraud Rate % by Hour)",
    labels={'Fraud_Rate_Percent': 'Fraud Percentage (%)'},
    color='Fraud_Rate_Percent',
    color_continuous_scale="Reds"
)


avg_rate = hourly_stats['Fraud_Rate_Percent'].mean()
fig.add_hline(y=avg_rate, line_dash="dash", line_color="blue",
              annotation_text=f"Average Risk: {avg_rate:.3f}%",
              annotation_position="top right")

fig.update_layout(template="plotly_white", height=500, showlegend=False)
fig.show()
```

The hourly fraud rate analysis reveals that certain low-volume hours (particularly 2–4 AM) show elevated fraud rates relative to the daily average. However, since the absolute number of transactions — and fraud cases — in these windows is small, these percentages carry high variance and should be interpreted with caution. This pattern is directionally interesting but requires validation on larger datasets before being operationalized as a threshold rule.

```python
# ==============================
# IMPORTS
# ==============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.preprocessing import RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline as SkPipeline
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score,
    precision_score,
    recall_score,
    f1_score,
    precision_recall_curve,
    auc,
    ConfusionMatrixDisplay
)

# ==============================
# DATA SPLIT (Train / Val / Test)
# ==============================
X = df.drop('Class', axis=1)
y = df['Class']

X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.25, stratify=y_temp, random_state=42
)

# ==============================
# PREPROCESSOR
# ==============================
preprocessor = ColumnTransformer(
    transformers=[
        ('robust_scaler', RobustScaler(), ['Amount', 'Time'])
    ],
    remainder='passthrough'
)

# ==============================
# PIPELINE
# ==============================
pipeline = SkPipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(class_weight='balanced', max_iter=1000))
])

# ==============================
# GRID SEARCH + CROSS VALIDATION
# ==============================
print("\n--- Grid Search + Cross Validation ---")

param_grid = {
    'classifier__C': [0.01, 0.1, 1, 10],
    'classifier__solver': ['liblinear', 'lbfgs']
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid_search = GridSearchCV(
    pipeline,
    param_grid,
    scoring='f1',
    cv=cv,
    n_jobs=1,
    verbose=1
)

grid_search.fit(X_train, y_train)

print("\nBest Parameters:")
print(grid_search.best_params_)

best_model = grid_search.best_estimator_

# ==============================
# THRESHOLD TUNING (Validation)
# ==============================
print("\n--- Threshold Tuning (Validation Set) ---")

y_val_proba = best_model.predict_proba(X_val)[:, 1]

thresholds = np.linspace(0.01, 0.99, 100)
results = []

for t in thresholds:
    y_pred_t = (y_val_proba >= t).astype(int)

    precision = precision_score(y_val, y_pred_t, zero_division=0)
    recall = recall_score(y_val, y_pred_t, zero_division=0)
    f1 = f1_score(y_val, y_pred_t, zero_division=0)

    results.append((t, precision, recall, f1))

results_df = pd.DataFrame(results, columns=['threshold', 'precision', 'recall', 'f1'])

best_row = results_df.loc[results_df['f1'].idxmax()]
optimal_threshold = best_row['threshold']

print("\nBest Threshold:")
print(best_row)

# ==============================
# FINAL TEST EVALUATION
# ==============================
print("\n--- Final Evaluation on Test Set ---")

y_test_proba = best_model.predict_proba(X_test)[:, 1]
y_pred_final = (y_test_proba >= optimal_threshold).astype(int)

print("\n[Confusion Matrix]")
print(confusion_matrix(y_test, y_pred_final))

print("\n[Classification Report]")
print(classification_report(y_test, y_pred_final))

print(f"\nROC-AUC Score: {roc_auc_score(y_test, y_test_proba):.4f}")

# ==============================
# PRECISION-RECALL CURVE
# ==============================
precision_curve, recall_curve, _ = precision_recall_curve(y_test, y_test_proba)
pr_auc = auc(recall_curve, precision_curve)

plt.figure()
plt.plot(recall_curve, precision_curve, label=f'PR AUC = {pr_auc:.4f}')

p_opt = precision_score(y_test, y_pred_final, zero_division=0)
r_opt = recall_score(y_test, y_pred_final, zero_division=0)

plt.scatter(r_opt, p_opt, label=f'Threshold = {optimal_threshold:.2f}')

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve (Final Model)")
plt.legend()
plt.show()

# ==============================
# CONFUSION MATRIX DISPLAY
# ==============================
ConfusionMatrixDisplay.from_predictions(
    y_test,
    y_pred_final,
    normalize='true',
    cmap='Blues',
    values_format='.2f'
)

plt.title("Final Model - Optimized Threshold")
plt.show()
```

```python
# ==============================
# 1. XGBOOST: TRAINING
# ==============================
from xgboost import XGBClassifier
from sklearn.pipeline import Pipeline as SkPipeline
from sklearn.model_selection import cross_val_score

print("--- 1. Training XGBoost ---")

neg_count = (y_train == 0).sum()
pos_count = (y_train == 1).sum()
spw = neg_count / pos_count

xgb_pipeline = SkPipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', XGBClassifier(
        learning_rate=0.05,
        n_estimators=200,
        max_depth=6,
        subsample=0.8,
        colsample_bytree=0.8,
        scale_pos_weight=spw,
        eval_metric='logloss',
        random_state=42,
        n_jobs=-1
    ))
])

xgb_pipeline.fit(X_train, y_train)

# ==============================
# Cross Validation
# ==============================
cv_scores = cross_val_score(
    xgb_pipeline,
    X_train,
    y_train,
    cv=5,
    scoring='roc_auc'
)

print(f"CV ROC-AUC: {cv_scores.mean():.4f}")


# ==============================
# 2. VALIDATION (Threshold Tuning)
# ==============================
print("\n--- Threshold Tuning (Validation) ---")

xgb_val_proba = xgb_pipeline.predict_proba(X_val)[:, 1]

thresholds = np.linspace(0.01, 0.99, 100)
results = []

for t in thresholds:
    y_pred_t = (xgb_val_proba >= t).astype(int)

    precision = precision_score(y_val, y_pred_t, zero_division=0)
    recall = recall_score(y_val, y_pred_t, zero_division=0)
    f1 = f1_score(y_val, y_pred_t, zero_division=0)

    results.append((t, precision, recall, f1))

results_df = pd.DataFrame(results, columns=['threshold', 'precision', 'recall', 'f1'])

best_row = results_df.loc[results_df['f1'].idxmax()]
optimal_threshold_xgb = best_row['threshold']

print("\nBest Threshold:")
print(best_row)


# ==============================
# 3. FINAL TEST EVALUATION
# ==============================
print("\n--- Final Evaluation (Test Set) ---")

xgb_test_proba = xgb_pipeline.predict_proba(X_test)[:, 1]
xgb_preds_opt = (xgb_test_proba >= optimal_threshold_xgb).astype(int)

print("\n[Classification Report - FINAL]")
print(classification_report(y_test, xgb_preds_opt))

print("\n[Confusion Matrix - FINAL]")
print(confusion_matrix(y_test, xgb_preds_opt))

print(f"\nROC-AUC Score: {roc_auc_score(y_test, xgb_test_proba):.4f}")


# ==============================
# 4. OVERFITTING CHECK
# ==============================
print("\n--- Overfitting Check ---")

train_proba = xgb_pipeline.predict_proba(X_train)[:, 1]
y_train_pred = (train_proba >= optimal_threshold_xgb).astype(int)

print("\n[Train Performance]")
print(classification_report(y_train, y_train_pred))

print("\n[Test Performance]")
print(classification_report(y_test, xgb_preds_opt))


# ==============================
# 5. PRECISION-RECALL CURVE
# ==============================
precision_curve, recall_curve, _ = precision_recall_curve(y_test, xgb_test_proba)
pr_auc = auc(recall_curve, precision_curve)

plt.figure()
plt.plot(recall_curve, precision_curve, label=f'PR AUC = {pr_auc:.4f}')

p_opt = precision_score(y_test, xgb_preds_opt, zero_division=0)
r_opt = recall_score(y_test, xgb_preds_opt, zero_division=0)

plt.scatter(r_opt, p_opt, label=f'Threshold = {optimal_threshold_xgb:.2f}')

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve (XGBoost)")
plt.legend()
plt.show()


# ==============================
# 6. CONFUSION MATRIX DISPLAY
# ==============================
ConfusionMatrixDisplay.from_predictions(
    y_test,
    xgb_preds_opt,
    normalize='true',
    cmap='Reds',
    values_format='.2f'
)

plt.title("XGBoost - Optimized Threshold")
plt.show()
```

1. Metric interpretation
The small differences in ROC-AUC between models are less informative in this highly imbalanced dataset. PR-AUC provides a more sensitive evaluation of minority class performance, where XGBoost achieves higher PR-AUC, indicating better fraud detection capability.

2. Threshold calibration
Logistic Regression requires a very high decision threshold (0.99), indicating limited class separability. XGBoost achieves optimal performance at a lower threshold (0.96), suggesting better probabilistic discrimination for fraud detection.

3. Model comparison
Under the current experimental setup, XGBoost achieves higher PR-AUC and F1-score compared to LightGBM, indicating better adaptation to the dataset’s imbalance and feature representation.

4. Production strategy
The high optimal threshold positions XGBoost as a precision-oriented fraud detection filter, prioritizing reduction of false positives while maintaining sufficient recall for operational deployment.
